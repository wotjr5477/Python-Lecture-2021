```python
# 연습문제
# 간단한 카운트 다운 프로그램 작성하여 보자. 60초부터 0초까지 숫자를 출력하고 0초 가량 되면 "발사"를 출력한다.
# 01. for문
for i in range(60, -1, -1):
    print(i, end=" ")
print("")
print("발사")

# 02. while문
num = 60
while num >=0:
    print(num, end=" ")
    num -=1
print("")
print("발사")

    

```
