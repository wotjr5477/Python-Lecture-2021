```python
# 마커 설정
A = np.random.standard_normal((100, 2))
B = np.random.standard_normal((100, 2))

A += np.array((-1, -2))
B += np.array((0.5, 1))
plt.scatter(A[:, 0], A[:, 1], color='k', marker='o')
plt.scatter(B[:, 0], B[:, 1], color='k', marker="^")
# ^ o x v > < s(스퀘어) p P * X D(다이아몬드)
plt.scatter(B[:, 0], B[:, 1], color='k', marker="x")
plt.scatter(B[:, 0], B[:, 1], color='k', marker="D")

plt.show()
```

```python
M = np.random.standard_normal((500, 2))
R = np.sum(M**2, axis=1)
plt.figure(figsize=(10, 10)) # figure를 지정하여 이미지 사이즈를 조정할 수 있다.
plt.ylim(-5, 5); plt.xlim(-5, 5) # lim을 주어 엣지에 여백을 줄 수 있다.
plt.scatter(M[:, 0], M[:, 1], c='0.7', marker='s',
            s=32*R, edgecolor='k')
plt.show()
```
